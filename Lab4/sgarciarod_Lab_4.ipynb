## SymPy’s Symbolic and Numercical Complex Evaluations
- https://docs.sympy.org/latest/modules/evalf.html](https://)
- x = x1 + I*x2,y = y1 + I*y2, z = z1 + I*z2, x * y * z

**My own Example 8**

imagen_2023-09-06_202000412.png

```
from sympy import symbols, I, N, pi, expand
x1, x2, x4, y1, y2, z1, z2, z3 = symbols("x1 x2 x4 y1 y2 z1 z2 z3", real=True)
x = x1/ (I*x2-x4)
y = y1 - I*y2
z = z1 * I*z2- z3

print(x*y*z)
print(expand(x*y*z))
print(expand((x*y)*z))
print(expand(x*(y*z)))
w=1/(pi + I)
print('w=',w)
w=N(w)
print('w=',w)
```
